### Setting a login in system using Loops

Blocking account after entering 3 successive wrong password

```python
# Set you initail Password
initial_password = input("Enter a New Password:")
print('Your password has been created successfully')

# creating a while loop to block accounts after 3 wrong attempts 
password = input("Enter your password:")
attempts = 1                                 # initial attempt
max_attempts = 3                             # maximum number of attepmts

# setting a while loop
while password != initial_password:
    if attempts < max_attempts:
        attempts += 1
        print("Wrong password. Enter you password again")
        password = input("Please enter your password again:")
    else:
        print("Contact support, your account has been block")
        break
else:
    print("Login successful. You are welcome")
```
